Download .json chat history of a mee6 channel using https://github.com/Tyrrrz/DiscordChatExporter

```python
import json
f = open("/content/Topcoder - --------___Games___-------- - token-commands [960608175164383302].json", encoding = "utf-8")
raw_data = json.load(f)
print(raw_data.keys())
```

```python
ranks = []
```

```python
for i in raw_data['messages']:
  if i['embeds']:
    if (i['embeds'][0]['title'])=='Topcoder Richest Players':
      ranks.append(i)
print(len(ranks))
```

```python
import pandas as pd
from datetime import datetime
rank_data = pd.DataFrame()
l = []
for i in ranks:
  try:
    dt = datetime.strptime((i['timestamp'].split('.')[0]),"%Y-%m-%dT%H:%M:%S")
  except:
    print((i['timestamp']))
    break
  fields = i['embeds'][0]['fields']
  total = int(fields[0]['value'].split()[1])
  data = {}
  for j in fields[1:4]:
    data[" ".join(j['name'].split()[1:])]=int(j['value'].split()[1])
  for j in fields[4:]:
    data[(j['name'].split('-')[1]).strip()]=int(j['value'].split()[1])
  data["datetime"]=dt
  # data['total']=total
  l.append(pd.DataFrame([data]))
final = pd.concat(l)
final = final.set_index(["datetime"])
# final = final.drop(['total'],)
final.head()
```

```python
# final.iloc[1:,:].head(100)
final = final.interpolate(method='linear',limit_direction="forward").fillna(0)
final.iloc[1:,:].head(100)
```
